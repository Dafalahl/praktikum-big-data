```python
# Task 1
from pymongo import MongoClient, InsertOne

client = MongoClient("mongodb://localhost:27017/")
db = client["university_db"]
courses_collection = db["courses"]

task1 = [
    InsertOne({
        'course': 'Math 101',
        'enrollments': 30,
        'department': 'Mathematics',
        'students': [
            {'name': 'Budi'},
            {'name': 'Yanto'},
            {'name': 'Bob'}
        ]
    }),
    InsertOne({
        'course': 'CS 102',
        'enrollments': 25,
        'department': 'Computer Science',
        'students': [
            {'name': 'Alex'},
            {'name': 'Bob'},
            {'name': 'Steve'}
        ]
    }),
    InsertOne({
        'course': 'History 201',
        'enrollments': 20,
        'department': 'History',
        'students': [
            {'name': 'Susi'},
            {'name': 'Santi'}
        ]
    }),
    InsertOne({
        'course': 'Physics 202',
        'enrollments': 15,
        'department': 'Physics',
        'students': [
            {'name': 'Max'}
        ]
    })
]

courses_collection.bulk_write(task1)
for course in courses_collection.find():
    print(course)

print("✅ Task 1 completed: courses inserted successfully!")
```

```python
# Query for courses with enrollments greater than 20
for course in courses_collection.find({'enrollments': {'$gt': 20}}):
    print(course)

```

```python
# Task 2
# Query courses in Computer Science or Mathematics departments
for course in courses_collection.find({'department': {'$in': ['Computer Science', 'Mathematics']}}):
    print(course)

print("\n")
print("✅ Task 2 completed: courses inserted successfully!")
```

```python
# Task 2
# Query courses in Computer Science or Mathematics departments
for course in courses_collection.find({'department': {'$in': ['Computer Science', 'Mathematics']}}):
    print(course)

print("\n")
# Query courses in Computer Science or Mathematics departments
for course in courses_collection.find({'department': {'$regex': 'Computer'}}):
    print(course)
    
print("\n")


for course in courses_collection.find({'department': {'$regex': 'computer', '$options': 'i'}}):
    print(course)

print("\n")
print("✅ Task 2 completed: courses inserted successfully!")
```

```python
pipeline = [
    {'$group': {'_id': '$department', 'average_enrollment':{'$avg':'$enrollments'}}}
]

for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Task 3
# Maximum enrollment per department
pipeline = [
    {'$group': {'_id': '$department', 'max_enrollment': {'$max': '$enrollments'}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)

print("✅ Task 3 completed: courses inserted successfully!")
```

```python
# Projection to rename fields
pipeline = [
    {'$project': {'course_name': '$course', 'department_name': '$department', 'enrollments': 1}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
```

```python
# Task 4
# Adding enrollment category field based on enrollments
pipeline = [
    {'$addFields': {'enrollment_category': {'$cond': {'if': {'$gt': ['$enrollments', 20]}, 'then': 'high', 'else': 'low'}}}}
]
for result in courses_collection.aggregate(pipeline):
    print(result)
# if you want more condition up then 2 you can use cond in else section or you can use switch
```

```python
# Homework 1
pipeline = [
    {
        '$group': {
            '_id': '$department',    
            'course_count': {'$sum': 1} 
        }
    },
    {
        '$sort': {'course_count': -1}  
    }
]

for result in courses_collection.aggregate(pipeline):
    print(result)
```
